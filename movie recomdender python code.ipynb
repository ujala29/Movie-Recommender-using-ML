```python
# i want [action, adventure, scince fiction]
import ast 
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        return l
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(4)
```

```python
def convert3(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
          l.append(i['name'])
          count+=1
        else:
         break
        return l
    
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

```python
# overwiew as string let convet intno list so that we can concantenate all in one tag
# Convert the string into a list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python
import ast

def convertdb(obj):
    try:
        crew_list = ast.literal_eval(obj)
        for member in crew_list:
            if member['job'] == 'Director':
                return member['name']
    except (ValueError, SyntaxError):
        return None

# Apply the function to the 'crew' column
movies['director'] = movies['crew'].apply(convertdb)
```

```python
# Drop the 'crew' column
movies = movies.drop('crew', axis=1)
movies.head(4)
```

```python
import ast

def safe_literal_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return val

# Convert the string representation of lists to actual lists
movies['genres'] = movies['genres'].apply(safe_literal_eval)

# Now apply the transformation to remove spaces
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
```

```python
movies['cast'] = movies['cast'].apply(safe_literal_eval)

# Now apply the transformation to remove spaces
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
```

```python
movies['keywords'] = movies['keywords'].apply(safe_literal_eval)

# Now apply the transformation to remove spaces
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
```

```python
# Convert each row of the 'director' column into a list
movies['director'] = movies['director'].apply(lambda x: [x] if x is not None else [])
```

```python
movies['director'] = movies['director'].apply(safe_literal_eval)

# Now apply the transformation to remove spaces
movies['director'] = movies['director'].apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else x)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
 vectors=cv.fit_transform(newdf['tags']).toarray()
```

```python
vectors[0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
stem('loving')
```

```python
 newdf['tags']=newdf['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movieindex=newdf[newdf['title']==movie].index[0]
    distance=similarity[movieindex]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
    
```

```python
recommend('The Dark Knight')
```
